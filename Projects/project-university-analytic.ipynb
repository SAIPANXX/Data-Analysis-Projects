```python
#step 4: conduct some statistical operations aka EDA = exploratory data analysis
df.describe()
# The commands returns statistical values such as mean, std, max, min 1st, 2nd, and 3rd quartile.
```

```python
highest_uni = df[df["Overall Score"] == df["Overall Score"].max()]
```

```python
lowest_rank
```

```python
country_counts = df["Country"].value_counts()
country_counts
```

```python
#Step 5: Data visualization
plt.figure(figsize = (10,6))
sns.barplot(x = country_counts.index,
           y = country_counts.values,
           palette = "viridis")
plt.title("The number of universities in each country")
plt.xlabel("Country")
plt.ylabel("No. of Universities")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Data visualisation: Scatter plot

plt.figure(figsize = (10,6))
sns.scatterplot(x = "Citations per Paper",
                y = "Overall Score",
                hue = "Country",
                data = df,
                palette = "deep")
plt.title("Relationship between Cications per Paper and Overall Score")
plt.xlabel("Citations per paper")
plt.ylabel("Overall Score")
plt.show()
```

```python
#Boxplot
plt.figure(figsize = (12,6))
sns.boxplot(x = "Country",
           y = "Overall Score",
           data = df,
           palette = "pastel")
plt.title("Coutry VS Overall Score Distribution")
plt.xlabel("Country")
plt.ylabel("Overall Score")
plt.xticks(rotation = 45)
plt.show()
```
