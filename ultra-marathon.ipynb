```python
# 1. differnce in speed for the 50k and 50 mi in male and female runners
df3.groupby(["race_length", "athlete_gender"])["athlete_average_speed"].mean()
```

```python
# 2. What age group are the best in the 50k race (20 + races min) (show 15)
df3.query('race_length == "50mi"').groupby("athlete_age")["athlete_average_speed"].agg(["mean", "count"]).sort_values("mean", ascending=False).query("count > 19").head(15)
```

```python
# what age group are the worst in the 50m race (10+ races min) (show 20)
df3.query('race_length == "50mi"').groupby("athlete_age")["athlete_average_speed"].agg(["mean", "count"]).sort_values("mean", ascending=True).query("count > 10").head(20)
```

```python
# season for the data -> slower in summer than winter?
"""
spring 3 - 5
summer 6 - 8
fall 9 - 11
winter 12 - 2

split between 2 decimals
"""
```

```python
df3["reace_month"] = df3["race_day"].str.split(".").str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3["race_season"] = df3["reace_month"].apply(lambda x: "Winter" if x > 11 else "Fal" if x > 8 else "Summer" if x > 5 else "Spring" if x > 2 else "Winter")
```

```python
df3.head(25)
```

```python
df3.groupby("race_season")["athlete_average_speed"].agg(["mean", "count"]).sort_values("mean", ascending = False)
```
